## Kernels and Filters

In image processing, a **kernel** is a **matrix of numbers**. In order to alter the image, we slide the kernel over it and apply **elementwise multiplication** and then a **sum**:

![convolution.png](convolution.png)
[(source)](https://anhreynolds.com/blogs/cnn.html)

```python
# Calculation above:
1*1 + 1*3 + 1*5 + 1*6 + 1*7 + 1*9
```

We then slide the kernel over by one pixel and apply the calculation again, and so on, until the whole image has been covered:

![kernel_moving.gif](kernel_moving.gif)
[(source)](https://towardsdatascience.com/intuitively-understanding-convolutions-for-deep-learning-1f6f42faee1)

This results in an **output image** that is **smaller than the original** and has been altered in some way. 

For multiple colour channels, a **kernel** is applied to **each colour channel** in the image. This collection of kernels is called a **filter**. Kernels are widely used to make image filters in image editing software, as well as for image feature extraction in deep learning, as we will see today. 

### Stride:

It is possible to move the kernel more than one pixel across at a time. This results in an even smaller output convolved image:

![stride.gif](stride.gif)
[(source)](https://towardsdatascience.com/intuitively-understanding-convolutions-for-deep-learning-1f6f42faee1)

### Padding: 

![padding.gif](padding.gif)
[(source)](https://towardsdatascience.com/intuitively-understanding-convolutions-for-deep-learning-1f6f42faee1)

When performing a convolution on an image, we end up losing the pixels on the edge.

However if we want to keep the outside pixels, we can use **padding**:
 - we add a single layer of pixels around the outside of the original image in order to extend it.
 - this new, slightly larger image is now convolved with the kernel
 - **the output is the same size as the original image**
 
The two most common kinds of padding are:
 - **same-value/mirror** - the fake new pixels have the same values as the ones they are next to in the original image
 - **zero** - the fake new pixels are set to zero 

The terms often seen in the context of CNNs in Keras is:
 - **no padding** - `padding = 'valid'` 
     - called **valid** because only valid pixels (i.e. those present in the original image) are used.
 - **zero padding** - `padding='same'` 
     - called **same** because the output has the same size as the input image (with stride=1)


### Pooling

![max_avg_pooling.png](max_avg_pooling.png)
[(source)](https://www.researchgate.net/publication/333593451_Application_of_Transfer_Learning_Using_Convolutional_Neural_Network_Method_for_Early_Detection_of_Terry's_Nail)

To make our feature space even smaller, we can reduce the convolved images further by **pooling**, where we take some numeric representation for subsets of the convolved image. This reduces the resolution of the image, and in fact makes it more robust against slight differences in locations of features. Pooling is also known as **subsampling**.

In the examples below we have:
- **max pooling** where we take 2x2 windows ($f=2$) of our convolved image, take the **maximum value of each** and **return an even smaller image**. We also have a **stride** of 2 ($s=2$), i.e. we don't have overlapping subsections.
    - max pooling ends up highlighting the most strongly present features in the sections of the image 
- **average pooling** where we take the **mean value** of each 2x2 window.
    - average pooling shows the presence of a feature on average in a section of the image

We can also add a bias to the convolution by adding some constant value to each pixel in the transformed image:

## Convolutional Neural Networks

In deep learning, we can use image convolutions to learn some features of an image (or set of images) that make. 


image adapted from [here](https://archana1998.github.io/post/rahul-sukthankar/)

### Convolution Layer(s)

Each of these layers has some number of kernels, whose values, along with the biases, are **learned by the network**.

**Example:** In the network above, the **number of parameters** for a first layer with **10 kernels** each with a **3x3 window** would be:
$$ (3\cdot3 + 1)\cdot 10 = 100$$

i.e. the network has to learn 100 parameters for that layer.


**N.B.** In the warmup, we were **choosing** kernels that did a **specific task** which was **human-interpretable** (edge detection, blurring, vertical edges, contrast etc).
It is important to understand that in the CNN, the kernels in the layers are determined **during training** through backpropagation. We do not tell the network what they are, it learns on its own! This can also mean that some features that it detects are not really interpretable by humans.

### Feature Maps

We add a bias to the pixels of the convolved image and apply an **activation function** (usually ReLu). The result is the output of our convolutional layer and is called a **feature map**:



[(source)](https://datahacker.rs/one-layer-covolutional-neural-network/)

In the image above we have a 3 colour channel input image, two filters (with three kernels each - one for each colour channel). The result (with stride = 1 and no padding) is a 4x4 image *for each filter*. 

We *add a bias* to each output image and apply the *ReLu activation function* (more on this later this afternoon) - this basically makes all pixels with negative values equal to 0 and all positive valued pixels stay the same.

We now have two 4x4 outputs showing different features of our image. We can stack them into a 4x4x2 array and this is the **feature map**. We can think about this as the set of features that the CNN "detects" or "highlights" from our image.

## CNN with Keras

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Dense, MaxPooling2D, Flatten
from tensorflow.keras import backend as K

from tensorflow.keras.optimizers import Adam

from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
```

### Preprocess the data:

```python
# Load data:
(Xtrain, ytrain),(Xtest, ytest) = mnist.load_data()
```

```python
# Check shapes:
Xtrain.shape, ytrain.shape, Xtest.shape, ytest.shape
```

```python
Xtrain[0].shape
```

```python
# We want to reshape our X data to the format (number, w, h, channels):
# MNIST is grayscale so there is just one colour channel
Xtrain = Xtrain.reshape(60000,28,28,1)
#equivalent reshaping for Xtest:
Xtest = Xtest.reshape(10000,28,28,1)
```

```python
# Always keep checking shapes:
Xtrain.shape,ytrain.shape,Xtest.shape,ytest.shape
```

```python
# Examine the data:
#Xtrain[0]
```

```python
# Have a look at the data graphically:
for i in range(25):
    plt.subplot(5, 5, i+1)
    plt.imshow(Xtrain[i], cmap='Greys')
    plt.axis('off')
    plt.text(0, 0, ytrain[i]) # displays y-values in each subplot
```

```python
#Can also check the target values:
ytrain[0]
```

```python
# Keep a copy of y test as the actual number labels:
ytest_true = ytest.copy()
```

```python
# One hot encode the labels:
ytrain = to_categorical(ytrain)
ytest = to_categorical(ytest)
```

```python
# See the difference:
ytrain[0]
```

### Make a model:

```python
# Clear session and build model:
K.clear_session()

model = Sequential([
    Conv2D(filters = 16,            # How many filters in this layer (n1 in the image above)
           kernel_size = (3,3),     # Kernel size
           strides = (1,1),         # Stride size in each dimension
           padding = "valid",       # Can also try "same" padding to keep image same size
           activation = "relu",     # More on this this afternoon!
           input_shape = (28,28,1)),# Very important! Input shape of image (H,W,colour_channels) 
    
    MaxPooling2D(pool_size = (2,2), # Specify what area to pool over
                 strides=(2,2),     # Stride for the pools
                 padding = "valid"),# Can also use "same" for an output same size as input 

    Conv2D(filters = 16, 
           kernel_size = (3,3), 
           strides = (1,1), 
           padding = "valid", 
           activation = "relu"),
    # Notice that we don't specify an input shape here, as it is taken care of by the network
    
    MaxPooling2D(pool_size = (2,2),
                 strides=(2,2),
                 padding = "valid"),
    
    Flatten(), #Unwind all our pixels into one long vector
    
    #Fully connected layers:
    
    Dense(50,activation='relu'), 
    
    #Output layer:
    Dense(10 ,activation= "softmax") 
    # same number of neurons as we have classes, softmax for multiclas classification

    
])
```

```python
# Check the summary:
model.summary()
```

```python
# Compile the model
model.compile(optimizer='Adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
# Fit the model
history = model.fit(Xtrain,ytrain,batch_size=1000, epochs=15,validation_split=0.2)
```

### Examine the model

```python
#history.history
```

```python
# Plot accuracy curve
plt.plot(history.history['accuracy'], label='training accuracy')
plt.plot(history.history['val_accuracy'], label= 'validation accuracy')
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.legend()

```

```python
# Plot loss
plt.plot(history.history['loss'], label='training loss')
plt.plot(history.history['val_loss'], label='validation loss')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.legend()
```

```python
ypred = model.predict(Xtest)
```

```python
# Look at the first prediction:
ypred[0]
```

```python
np.argmax(ypred[0])
```

```python
# How do we interpret our predictions?
plt.imshow(Xtest[0],cmap='gray')
```

```python
ytest_true[0], ytest[0]
```

```python
# Plot confusion matrix
cm = confusion_matrix(y_true=ytest_true, y_pred=np.argmax(ypred, axis=1))

disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                             display_labels=np.unique(ytest_true))
fig, ax = plt.subplots(figsize=(10,10))

disp.plot(ax=ax)

# If this doesn't work, you have an old version of sklearn and need to upgrade it! 

```

## Bonus: How to visualise feature maps

```python
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.image import img_to_array
from numpy import expand_dims
```

We can access the model layers using the `.layers` attibute:

```python
model.layers
```

We can also access the `inputs` attibute:

```python
model.inputs
```

We can also build a new redefined mini-model that takes the same inputs, and outputs something immediately after the first convolutional layer:

```python
# redefine model to output right after the first hidden layer
K.clear_session()
model_small = Model(inputs=model.inputs, outputs=model.layers[0].output)
model_small.summary()
```

We can load an image with the required shape:


```python
img = Xtrain[2]
plt.imshow(img.reshape(28,28), cmap=plt.cm.Greys)
```

```python
img.shape
```

Our image is of the shape (28,28,1) and we want it to correspond to the input shape of (1,28,28,1) (one slice of toast :) ) 

```python
# expand dimensions so that it represents a single 'sample'
img=img.reshape(1,28,28,1)

#equivalent to
#img = expand_dims(img, axis=0)

#(generally there are lots of ways of doing this in numpy!)

img.shape

```

```python
# get feature map for first hidden layer
feature_maps = model_small.predict(img)
```

```python
# plot all 16 maps:
height = 4
width = 4
ix = 1
plt.figure(figsize=(8,8))
for _ in range(height):
    for _ in range(width):
        # specify subplot and turn of axis
        ax = plt.subplot(height, width, ix)
        ax.set_xticks([])
        ax.set_yticks([])
        # plot filter channel in grayscale
        plt.imshow(feature_maps[0, :, :, ix-1], cmap='gray')
        ix += 1
# show the figure
plt.show()
```

Similarly we can look at the output of the first max pooling layer by changing the output to `model.layers[1]` above

```python
K.clear_session()
model_small_pool = Model(inputs=model.inputs, outputs=model.layers[1].output)
model_small_pool.summary()
```

```python
feature_maps_pool = model_small_pool.predict(img)
# plot all 16 maps:
height = 4
width = 4
ix = 1
plt.figure(figsize=(8,8))
for _ in range(height):
    for _ in range(width):
        # specify subplot and turn of axis
        ax = plt.subplot(height, width, ix)
        ax.set_xticks([])
        ax.set_yticks([])
        # plot filter channel in grayscale
        plt.imshow(feature_maps[0, :, :, ix-1], cmap='gray')
        ix += 1
# show the figure
plt.show()
```
