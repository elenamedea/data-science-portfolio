```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("customer_data.csv", parse_dates = True, index_col = 0)
```

```python
df.head()
```

```python
df.sort_values(by = ["timestamp", "customer_id"], inplace = True)
```

```python
df
```

```python
df = df.groupby(by=["customer_id"]).resample(rule='T').last()
df.head()
```

```python
df = df.fillna(method = "ffill")
```

```python
df.head()
```

```python
df.rename(columns = {"location" : "before"}, inplace = True)
df.head()
```

```python
df["after"] = df["before"]
df.head()
```

```python
df["before"] = df["before"].shift(1)
df.head()
```

```python
df["before"].unique()
```

```python
df["before"].replace({"checkout" : "entrance"}, inplace = True)
```

```python
df["before"].fillna("entrance", inplace = True)
```

```python
transition_frequencies = pd.crosstab(df["before"], df["after"], normalize = 0)
transition_frequencies
```

```python
transition_frequencies.sum(axis = 1)
```

```python
transition_frequencies["exit"] = [0, 0, 0, 0, 0]
```

```python
checkout = {"checkout" : [0], "dairy" : [0], "drinks" : [0], "fruit" : [0], "spices" : [0], "exit" : [1]}
```

```python
checkout = pd.DataFrame(data = checkout)
```

```python
checkout.rename(index = {0 : "checkout"}, inplace = True)
```

```python
checkout
```

```python
transition_frequencies = pd.concat([transition_frequencies, checkout])
```
